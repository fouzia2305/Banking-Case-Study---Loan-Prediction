### Check for outliers

```python
def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 *IQR)
    outliers_below=column[column<lower_bound]
    outliers_above=column[column>upper_bound]
    outliers = column[(column < lower_bound) | (column > upper_bound)]

    outliers_count=len(outliers_below)+len(outliers_above)
    return   outliers,lower_bound,upper_bound,len(outliers_below),len(outliers_above),outliers_count
```

```python
for col in df_num_cols:
    outliers,lower_bound,upper_bound,outliers_below,outliers_above,outliers_count=find_outliers_iqr(df[col])
    print(f'column:{col}')
    print(f'lower_bound:{lower_bound}')
    print(f'upper_bound:{upper_bound}')
    print(f'Outliers count below lower bound:{outliers_below}')
    print(f'Outliers count above upper bound:{outliers_above}')
    print(f'Total Outliers Count :{outliers_count}')
    print('----------------------------------------------------')
```

```python
def remove_outliers(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
def replace_higher_outliers(df, col):
    Q3 = df[col].quantile(0.75)
    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
    upper_boundary = Q3 + (1.5 * IQR)

    df.loc[df[col] > upper_boundary, col] = upper_boundary

    return df


df = replace_higher_outliers(df, 'person_age')
df = replace_higher_outliers(df, 'person_income')
df = replace_higher_outliers(df, 'person_emp_length')
df = replace_higher_outliers(df, 'loan_amnt')
df = replace_higher_outliers(df, 'loan_int_rate')
df = replace_higher_outliers(df, 'loan_percent_income')
df = replace_higher_outliers(df, 'person_income')
df = replace_higher_outliers(df, 'cb_person_cred_hist_length')

```
